```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

Less = data.loc[data['PREAUTH_AMT'] < data['CLAIM_AMOUNT']].count()[0]
equal = data.loc[data['PREAUTH_AMT'] == data['CLAIM_AMOUNT']].count()[0]
more = data[data['PREAUTH_AMT'] > data['CLAIM_AMOUNT']].count()[0]


comparision = [ Less , equal , more]
label = [ "PRE_AUTH < CLAIM" , "PRE_AUTH = CLAIM" , "PRE_AUTH > CLAIM" ]

explode = (0,0.2,0.2)

plt.title('COMPARISION BETWEEN CLAIM AND PRE_AUTH AMOUNT')

plt.pie(comparision, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()

plt.savefig('claim and pre-auth comparisons.png')
```

```python
fig = plt.gcf()
fig.set_size_inches(10,5
                )
fatalities = pd.crosstab(data['SURGERY'],data['Mortality Y / N'])
fatalities.sort_values(by=['YES'],ascending=False)[:10]['YES'].plot.bar(title= 'Surgeries with Most Fatalities')

plt.ylabel("Number of fatalities")
plt.show()
```

```python
fig = plt.gcf()
fig.set_size_inches(10,7)
norm_fatalities = fatalities
norm_fatalities['YES_scaled'] = norm_fatalities['YES']/(norm_fatalities['YES']+norm_fatalities['NO'])*100
norm_fatalities.sort_values(by=['YES_scaled'],ascending=False )[:10]['YES_scaled'].plot.bar(title= 'Most fatal surgeries' , color = "darkgoldenrod")
```
